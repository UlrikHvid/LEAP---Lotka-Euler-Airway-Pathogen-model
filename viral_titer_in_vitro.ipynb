```python
import numpy as np
import matplotlib.pyplot as plt

orange = "#fa6100"
```

### Data read off from https://pmc.ncbi.nlm.nih.gov/articles/PMC8650692/pdf/fcimb-11-771524.pdf using graphreader.com

```python
t_delta = np.array([2,4,6,8,10,12,24,48])
titer_delta = np.array([
    104.1721,
    249.1436,
    3599.2583,
    9090.3349,
    100000,
    2491435.8201,
    8726265.0969,
    1444652.2343,
])

t_wt = np.array([2,4,6,8,16,24,36,48,72])

titer_wt = np.array([
    154.3142,
    188.5221,
    343.7414,
    10339.3385,
    222754.2952,
    791681.9159,
    1763505.472,
    5484416.5761,
    3215484.177,
])

plt.figure(figsize=(10,5))
plt.plot(t_delta, titer_delta,color = orange)
plt.title("Delta",fontsize = 20)
plt.yscale("log")
plt.grid()

plt.figure(figsize=(10,5))
plt.plot(t_wt, titer_wt,color = orange)
plt.title("Wild-type",fontsize = 20)
plt.yscale("log")
plt.grid()
```

```python
Lamb = 0.25
titer_eff_delta = titer_delta * np.exp(-t_delta*Lamb)
titer_eff_wt = titer_wt * np.exp(-t_wt*Lamb)

plt.figure(figsize=(10,5))
plt.plot(t_delta,titer_eff_delta, color = orange, linewidth = 2,marker = "o")
#plt.title("Delta, effective reproduction",fontsize = 20)
plt.xticks([2,4,6,8,10,12,16,24,36,48,72],fontsize = 12)
plt.xlabel("Time (hpi)",fontsize = 20)
plt.ylabel("Discounted titer",fontsize = 20)
plt.xlim(1,75)
#plt.yticks(fontsize = 0)
#plt.yscale("log")
plt.grid()

plt.figure(figsize=(10,5))
plt.plot(t_wt, titer_eff_wt, color = orange, linewidth = 2,marker = "o")
plt.xticks([2,4,6,8,10,12,16,24,36,48,72],fontsize = 12)
#plt.title("Wild-type, effective reproduction",fontsize = 20)
#plt.yticks(fontsize = 0)
plt.xlabel("Time (hpi)",fontsize = 20)
plt.ylabel("Discounted titer",fontsize = 20)
plt.xlim(1,75)
#plt.yscale("log")
plt.grid()
```
